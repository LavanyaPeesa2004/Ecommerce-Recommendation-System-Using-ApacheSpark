```python
silver_df.createOrReplaceTempView("silver")
```

# Top 20 Products

```python
display(
spark.sql("""
SELECT product_name,
       COUNT(*) AS total_purchases
FROM silver
GROUP BY product_name
ORDER BY total_purchases DESC
LIMIT 20
""")
)
```

## Top 10 Departments

```python
display(
    spark.sql("""
    select department,
    count(*) as total_orders 
    from silver 
    group by department 
    order by total_orders desc
    limit 10 
    """)
)
```

## Top 10 Aisles

```python
display (
    spark.sql("""
    select aisle ,
    count(*) as total_orders
    from silver
    group by aisle
    order by total_orders desc 
    limit 10          
  """)
)
```

## Percentage Of Reordered

```python
display(
spark.sql("""
SELECT
AVG(reordered)*100 AS reorder_percentage
FROM silver
""")
)
```

## Shopping By hours

```python
display(
spark.sql("""
SELECT order_hour_of_day,
COUNT(*) AS total_orders
FROM silver
GROUP BY order_hour_of_day
ORDER BY order_hour_of_day
""")
)
```

## Loyal_Customers

```python
display (
    spark.sql("""
    select user_id,
    count (*) as total_purchases
    from silver 
    group by user_id
    order by total_purchases desc
    limit 20 
    """
    )
)
```

## Shopping By Week

```python
display(
    spark.sql("""
    select order_dow,
    count(*)
    from silver
    group by order_dow
    order by order_dow desc
    """)
)
```
